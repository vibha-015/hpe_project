```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('C://Users//Admin//OneDrive//Desktop//HPE//data_1.csv')
```

```python
data.shape
```

```python
data['START_TIME'] = pd.to_datetime(data['START_TIME'])
```

```python
data['tDiff'] = data.START_TIME.diff()
```

```python
data[data.tDiff > pd.Timedelta('1h')]
```

```python
df=data
```

```python
df=df.set_index('START_TIME').resample('H').mean()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.ffill(axis = 0)
```

```python
df.isnull().sum()
```

```python
cor=df.corr()
cor
```

```python
plt.figure(figsize=(25,20))
sns.heatmap(data=cor,annot=True,cmap='mako')
plt.savefig(f"heatmap.png")
```
