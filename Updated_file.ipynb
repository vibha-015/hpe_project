```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('C://Users//Admin//OneDrive//Desktop//HPE//data_1.csv')
```

### HEATMAP

```python
cor=data.corr()
plt.figure(figsize=(30,25))
sns.heatmap(data=cor,annot=True,cmap='coolwarm')
plt.savefig(f"heatmap.png")
```

### BOXPLOTS

```python
data.plot(kind='box',subplots=True,figsize=(50,200),color='red',layout=(13,3),ylabel="values")
plt.savefig(f"boxplot.png")
```

### REGRESSION PLOTS

```python
sns.lmplot(x="F11", y="F20", data=data)
plt.title("Regression plot between F11 and F20")
plt.xlabel("F11")
plt.ylabel("F20")
plt.show()
```
